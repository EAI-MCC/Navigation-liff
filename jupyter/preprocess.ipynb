```python
%matplotlib notebook
trajectory = [111, 107, 103, 102, 82, 62, 42, 43, 39, 35, 34, 35, 34, 35, 34, 35, 34, 35, 34, 35, 34, 35, 34, 35, -1, 34, 35, -1, -1, 34, 35, 34, 2, 34, 54, 34, 35, 34, 35, -1, 34, 35, 34, 35, 34, 35, 34, 35, 34, 35, 34, 2, 34, 35, 34, 35, 34, 35, 34, 35, 34, 35, 34, 35, 34, 35, 34, 2, 34, 2, 34, 35, -1, 34, 35, 34, 35, 34, 35, 34, 35, 34, 35, 34, 35]
with h5py.File('../data/'+'bedroom_06.h5') as hf:
    data_np_array = np.array(hf['location'])
    rotation_np_array = np.array(hf['rotation'])
    image_np_array = np.array(hf['observation'])
    graph_np_array = np.array(hf['graph'])
    data_x = data_np_array[::,0]
    data_y = data_np_array[::,1]
    fig = plt.figure()
    ax = fig.add_subplot(111)
    plt.ion()
    plt.scatter(data_x,data_y)
    plt.ylim(-1,9)
    plt.xlim(-1,10)
    fig.show()
    fig.canvas.draw()
    for id in trajectory:
        time.sleep(0.5)
        print(id,'  graph -> ',graph_np_array[id],' location -> ',data_np_array[id])
        ax.clear()
        plt.ylim(-1,9)
        plt.xlim(-1,10)
        current_x =  data_np_array[id,0]
        current_y =  data_np_array[id,1]
        if id == -1:
            ax.scatter(data_x,data_y,c = 'r')
            fig.canvas.draw()
        else:
            angle = rotation_np_array[id]
            plt.scatter(data_x,data_y,c = 'b')
            plt.scatter(current_x,current_y,c = 'r')
            dx = 0.5*np.cos(np.deg2rad(90-angle))
            dy = 0.5*np.sin(np.deg2rad(90-angle))
            plt.scatter(current_x+dx,current_y+dy,c = 'g')
            fig.canvas.draw()
```

```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt

m = 100
n = 100
matrix = np.random.normal(0,1,m*n).reshape(m,n)

fig = plt.figure()
ax = fig.add_subplot(111)
plt.ion()

fig.show()
fig.canvas.draw()

for i in range(0,100):
    ax.clear()
    ax.plot(matrix[i,:])
    fig.canvas.draw()
```
